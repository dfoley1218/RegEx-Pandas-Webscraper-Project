```python
from bs4 import BeautifulSoup
import requests
```

```python
# Here we are going to take a website with scraped text and determine a word count from a specific body
# We will use MLK's famous speech below to attempt this
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('p')
```

```python
mlkj = soup.find_all('p')
```

```python
mlkj
```

```python
# let's remove all of the P tags from the body of text. 
```

```python
type(mlkj)
```

```python
speech_combined = [p.text for p in mlkj]

print(speech_combined)
```

```python
# we are still seperated by commas and the \r\n. This needs to be fixed.
```

```python
speech_combined = [p.text for p in mlkj]

type(speech_combined)
```

```python
# we don't want all of these items to be in a list, we want them to be together
```

```python
' '.join(speech_combined)
```

```python
string_speech = ' '.join(speech_combined)
```

```python
# now it is just one long string
```

```python
string_speech[:100]
```

```python
# let's now see if we can remove the \r\
```

```python
string_speech.replace('\r\n', ' ')
```

```python
cleaned_speech = string_speech.replace('\r\n', ' ')
```

```python
# now let's get rid of those \'s bits'

# we are now going to just remove all punctuation altogether, we won't need it for a word count
```

```python
import re
```

```python
re.sub(r'[^\w\s]', '',cleaned_speech)
```

```python
speech_punc = re.sub(r'[^\w\s]', '',cleaned_speech)
```

```python
speech_lower = speech_punc.lower()
```

```python
# now we want to split all of the words into their own values, seperated by the words themselves.
# we can use the split method with regex into a list
```

```python
re.split(r'\s+', speech_lower)
```

```python
speech_broken = re.split(r'\s+', speech_lower)
```

```python
import pandas as pd
```

```python
pd.DataFrame(speech_broken).value_counts()
```

```python
df = pd.DataFrame(speech_broken).value_counts()
df
```
